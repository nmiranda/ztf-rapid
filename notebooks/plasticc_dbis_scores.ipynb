# Class Balancing: Results with Synthetic Data

First run `

```python
import numpy as np
from tensorflow.keras.models import load_model
from ztfrapid import ztf_rapid
from sklearn import metrics
```

```python
files = np.load('/home/miranda/ztf-rapid/data/processed/test_plasticc_none/test_plasticc_none.npz')
```

```python
model = load_model('/home/miranda/ztf-rapid/models/test_plasticc_none/test_plasticc_none.hdf5')
```

```python
y_pred = ztf_rapid.predict(model, files['X_test'])
y_pred = ztf_rapid.get_pred_label(y_pred)
```

```python
y_true = ztf_rapid.get_y_true(files['y_test'])
```

```python
ztf_rapid.plot_confusion_matrix(y_true, y_pred, files['class_names'])
```

```python
print(metrics.classification_report(y_true, y_pred, target_names=files['class_names']))
```
