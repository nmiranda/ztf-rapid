```python
import numpy as np
from ztfrapid import ztf_rapid
from tensorflow.keras.models import load_model
```

```python
files = np.load('/home/nmiranda/workspace/ztf_rapid/data/processed/test_none/test_none.npz')
X_test = files['X_test']
y_test = files['y_test']
class_names = files['class_names']
list(files.keys())
```

```python
model_filepath = '/home/nmiranda/workspace/ztf_rapid/models/test_none/test_none.hdf5'
model = load_model(model_filepath)
```

```python
y_pred = ztf_rapid.predict(model, X_test)
pred_label = ztf_rapid.get_pred_label(y_pred, time_index=10)
y_true = ztf_rapid.get_y_true(y_test)
ztf_rapid.plot_confusion_matrix(y_true, pred_label, class_names)
```

```python
y_pred = ztf_rapid.predict(model, X_test)
pred_label = ztf_rapid.get_pred_label_peak(X_test, y_pred)
y_true = ztf_rapid.get_y_true(y_test)
ztf_rapid.plot_confusion_matrix(y_true, pred_label, class_names)
```

```python
y_pred = ztf_rapid.predict(model, X_test)
pred_label = ztf_rapid.get_pred_label(y_pred, time_index=-1)
y_true = ztf_rapid.get_y_true(y_test)
ztf_rapid.plot_confusion_matrix(y_true, pred_label, class_names)
```
