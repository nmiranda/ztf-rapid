```python
import numpy as np
from tensorflow.keras.models import load_model
from ztfrapid import ztf_rapid
```

```python
datasets = np.load('/home/miranda/ztf-rapid/data/processed/test_noisify/test_noisify.npz', allow_pickle=True)
```

```python
model = load_model('/home/miranda/ztf-rapid/models/test_noisify/test_noisify.hdf5')
```

```python
list(datasets.keys())
```

```python
y_pred = ztf_rapid.predict(model, datasets['X_test'])
y_pred = ztf_rapid.get_pred_label(y_pred)
y_true = ztf_rapid.get_y_true(datasets['y_test'])
```

```python
ztf_rapid.plot_confusion_matrix(y_true, y_pred, datasets['class_names'])
```
