```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from ztfrapid import ztf_rapid
```

```python
files = np.load('/home/miranda/ztf-rapid/data/processed/test_over_001.npz')
```

```python
fluxes = np.ravel(files['X_test'])
mask = fluxes > 0.0
fluxes = fluxes[mask]
# mask = fluxes < 10000.0
# fluxes = fluxes[mask]

plt.figure(figsize=(14,6))
plt.yscale('log')
_, edges, _ = plt.hist(fluxes, bins=20)
# ticks = plt.xticks(np.linspace(start=0.0, stop=edges[-1], num=20))
ticks = plt.xticks(np.arange(start=0.0, stop=edges[-1], step=2500))
```

```python
model = load_model('/home/miranda/ztf-rapid/models/test_over_100/test_over/test_over_001.hdf5')
```

```python
X_test_brightest, y_test_brightest = ztf_rapid.select_bright_sources(files['X_test'], files['y_test'], 500.0)
print(files['X_test'].shape[0])
print(X_test_brightest.shape[0])
y_pred_brightest = ztf_rapid.predict(model, X_test_brightest)
y_pred_brightest = ztf_rapid.pred_to_categorical(y_pred_brightest)
y_test_brightest = ztf_rapid.target_to_categorical(y_test_brightest)
ztf_rapid.plot_confusion_matrix(y_test_brightest, y_pred_brightest, files['class_names'])
```
